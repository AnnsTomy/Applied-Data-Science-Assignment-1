```python
"""Anns Tomy"""
"""Student I D: 22033815"""
#importing the required libraries for plotting and reading the given excel file.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import textwrap
```

```python
"""read the excel file and demonstrate it as a dataframe"""
df_life_expectancy =  pd.read_excel("POPULAR INDICATORS LIFE EXPECTANCY.xlsx", index_col = 0)
df_life_expectancy
```

```python
#dropping the unwanted columns
df_life_expectancy1 = df_life_expectancy.drop(["Unnamed: 11"], axis = 1)
df_life_expectancy1
```

```python
#slicing the dataframe 
df_life_expectancy2 = df_life_expectancy1.iloc[:,0:9]
df_life_expectancy2
```

```python
# plot the Life expectancy rate using lineplot
plt.figure()
plt.plot(df_life_expectancy2.columns, df_life_expectancy2.loc["Albania"], label = "Albania")
plt.plot(df_life_expectancy2.columns, df_life_expectancy2.loc["Algeria"], label = "Algeria")
plt.plot(df_life_expectancy2.columns, df_life_expectancy2.loc["Afghanistan"], label = "Afghanistan")
plt.plot(df_life_expectancy2.columns, df_life_expectancy2.loc["Aruba"], label = "Aruba")
plt.legend()
plt.xlabel("Years")
plt.ylabel("Life Expectancy at birth")
plt.title("Life Expectancy")
plt.show()
```

```python
#plotting the Life expectancy rate of the year 2000 using barplot
plt.figure(figsize = (16,8))
plt.bar(df_life_expectancy2.index, df_life_expectancy2["2000"])



plt.xlabel("counries")
plt.ylabel("Life Expectancy at birth")
plt.title("Life Expectancy rate of the year 2000")
plt.show()
```
